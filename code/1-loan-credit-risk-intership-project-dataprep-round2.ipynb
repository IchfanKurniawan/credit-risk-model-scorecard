### Weight of Evidence & Value Information

```python
def woe_discrete_data(X_train, col, y_train):
    temp_df = pd.concat([X_train[col], y_train], axis=1)
    cols = temp_df.columns.values

    temp_df = pd.concat([
                            temp_df.groupby(cols[0])[cols[1]].count(),
                            temp_df.groupby(cols[0])[cols[1]].mean()
                        ], axis=1)

    temp_df.reset_index(inplace=True)
    temp_df.columns = [cols[0], 'n_obs', 'prob_good']
    temp_df['prob_n_obs'] = temp_df['n_obs'] / temp_df['n_obs'].sum()

    temp_df['n_good'] = temp_df['n_obs'] * temp_df['prob_good']
    temp_df['n_bad'] = temp_df['n_obs'] - temp_df['n_good']

    temp_df['prob_n_good'] = temp_df['n_good']/temp_df['n_good'].sum()
    temp_df['prob_n_bad'] = temp_df['n_bad']/temp_df['n_bad'].sum()

    temp_df['woe'] = np.log(temp_df['prob_n_good'] / temp_df['prob_n_bad'])
    temp_df.sort_values('woe', inplace=True)

    temp_df['diff_prob_good'] = temp_df['prob_good'].diff().abs()
    temp_df['diff_woe'] = temp_df['woe'].diff().abs()

    temp_df['info_value'] = temp_df['woe'] * (temp_df['prob_n_good'] - temp_df['prob_n_bad'])
    temp_df['info_value'] = temp_df['info_value'].sum()
    
    return temp_df
```

```python
def woe_conti_data(X_train, col, y_train):
    temp_df = pd.concat([X_train[col], y_train], axis=1)
    cols = temp_df.columns.values

    temp_df = pd.concat([
                            temp_df.groupby(cols[0])[cols[1]].count(),
                            temp_df.groupby(cols[0])[cols[1]].mean()
                        ], axis=1)

    temp_df.reset_index(inplace=True)
    temp_df.columns = [cols[0], 'n_obs', 'prob_good']
    temp_df['prob_n_obs'] = temp_df['n_obs'] / temp_df['n_obs'].sum()

    temp_df['n_good'] = temp_df['n_obs'] * temp_df['prob_good']
    temp_df['n_bad'] = temp_df['n_obs'] - temp_df['n_good']

    temp_df['prob_n_good'] = temp_df['n_good']/temp_df['n_good'].sum()
    temp_df['prob_n_bad'] = temp_df['n_bad']/temp_df['n_bad'].sum()

    temp_df['woe'] = np.log(temp_df['prob_n_good'] / temp_df['prob_n_bad'])

    temp_df['diff_prob_good'] = temp_df['prob_good'].diff().abs()
    temp_df['diff_woe'] = temp_df['woe'].diff().abs()

    temp_df['info_value'] = temp_df['woe'] * (temp_df['prob_n_good'] - temp_df['prob_n_bad'])
    temp_df['info_value'] = temp_df['info_value'].sum()
    
    return temp_df
```

```python
def plot_woe(col, rotation = 90):
    plt.figure(figsize=(15,6))
    sns.lineplot(x=temp_df[col], y=temp_df['woe'], color='gray', marker = 'o', linestyle = '--',)
    plt.xlabel(col)
    plt.ylabel('woe')
    plt.xticks(rotation = rotation)
    plt.show()
```

```python
def plot_woe_conti(col, rotation = 90):
    plt.figure(figsize=(15,6))
    sns.lineplot(x=temp_df.index, y=temp_df['woe'], color='gray', marker = 'o', linestyle = '--',)
    plt.xlabel(col)
    plt.ylabel('woe')
    plt.xticks(rotation = rotation)
    plt.show()
```

### Numerical Features

```python
# multicollinearity check

target = 'target'
threshold = .75
drop_cols = []

bulk_corr = df.corr()
corr_target = bulk_corr[target].reset_index().sort_values(target, ascending=False) # correlation all features w/ target
corr_all = bulk_corr.drop(labels=target, axis=1).drop(labels=target, axis=0) # correlation all feature w/ target

for col in corr_all.columns:
    if col in drop_cols:
        continue
    
    else:
        multi = corr_all[corr_all[col] > threshold][col]
        indexes_feats = multi.index
        
        if len(indexes_feats) > 0:
            for feat in indexes_feats:
                if feat == col:
                    continue
                else:
                    col_corr = abs(corr_target[corr_target['index'] == col]['target'].values[0]) # correlation of col & target
                    feat_corr = abs(corr_target[corr_target['index'] == feat]['target'].values[0])  # correlation of feat & target

                    if col_corr > feat_corr:
                        drop_cols.append(feat)
                    else:
                        drop_cols.append(col)
                        break
                        
drop_cols = np.unique(np.array(drop_cols))
```
